## Exercise 4 :  A function for Fibonacci

```python
def fibonacci_func(n):
    if n <= 0:
        return "Error!"
    if n == 1:
        return [0]
    fibonnaci = [0,1]
    for i in range(0,n-2,1):
        fibonnaci = fibonnaci + [fibonnaci[-2] + fibonnaci[-1]]
    return fibonnaci

fibonacci_func(10)
```

```python
def fibonacci_func(n):
    if n <= 0:
        return "Error!"
    if n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    fibonnaci = [0,1]
    for i in range(0,n-2,1):
        fibonnaci = fibonnaci + [fibonnaci[-2] + fibonnaci[-1]]
    return fibonnaci

fibonacci_func(2)
```

## Exercise 5 :  max_info() without NumPy

```python
numbers_1 = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
```

```python
def max_info(numbers):
    my_max = numbers[0]
    for i in numbers: 
        if i > my_max:
            my_max = i
            my_max_index = numbers.index(my_max)
            return my_max, my_max_index
    
max_info(numbers)
```
