## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
```

```python
plt.plot(x, y_1, color = 'red', linestyle = 'dashed', label = 'Y values')
# shortcut: color = c, linestyle = ls, linewidth = lw
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 0.25)
plt.legend(loc = 'upper left')
plt.show()

# plt.plot(x, y) to keep it simple
```

### 1.2 Adding another plot

```python
plt.plot(x, y_1, color = 'red', linestyle = 'dashed', label = 'Y values')
plt.plot(x, y_2, c = 'blue', label = 'Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 0.25)
plt.legend(loc = 'upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 0.1, 0.75
```

```python
plt.plot(x, y_1, c = 'red', ls = 'dashed', label = 'Y values')
plt.plot(x, y_2, c = 'blue', label = 'Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error,
            c = 'green', label = 'Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 0.25)
plt.legend(loc = 'upper left')
plt.show()
```

### 1.4 Experiment, Experiment, Experiment

```python
plt.plot(x, y_1, c = 'red', ls = 'dashed', label = 'Y values')
plt.plot(x, y_2, c = 'blue', label = 'Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error,
            c = 'green', label = 'Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 1, c = 'red', ls = 'dashed') # increases the transparency of the grids
plt.legend(loc = 'upper left')
plt.show() 
```

```python
plt.plot(x, y_1, c = '#dd1c77', ls = 'dashed', label = 'Y values') # can put the hex number to change color
plt.plot(x, y_2, c = 'blue', label = 'Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error,
            c = 'green', label = 'Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 1, c = 'red', ls = 'dashed') # increases the transparency of the grids
plt.legend(loc = 'upper right') # without specifying, it is still on the upper left
plt.show() 
```

### 1.5 Better with NumPy

```python
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c = 'red', ls = 'dashed', label = 'Y values')
plt.plot(np_x, np_y_2, c = 'blue', label = 'Y2 values')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x ** 2
sin_x = np.sin(x)

plt.plot(x, x2, c = 'red', ls = 'dashed', label = 'x^2')
plt.plot(x, sin_x, c = 'blue', label = 'sin(x)')
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x ** 2
sin_x = np.sin(x)

plt.plot(x, x2, c = 'red', ls = 'dashed', label = 'x^2')
plt.legend(loc = 'lower left')

plt.twinx() # adding new axis on the right
plt.plot(x, sin_x, c = 'blue', label = 'sin(x)')
plt.legend(loc = 'lower right')

plt.show()
```

```python
x = np.linspace(0, 6, 100)
x2 = x ** 2
sin_x = np.sin(x)

plt.plot(x, x2, c = 'red', ls = 'dashed', label = 'x^2')
plt.legend(loc = 'lower left')

plt.twinx() # adding new axis on the right
plt.plot(x, sin_x, c = 'blue', label = 'sin(x)')
plt.legend(loc = 'lower right')

plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x ** 2
sin_x = np.sin(x)

plt.plot(x, x2, c = 'red', ls = 'dashed', label = 'x^2')
plt.legend(loc = 'lower left')

plt.twinx()
plt.plot(x, sin_x, c = 'blue', label = 'sin(x)')
plt.legend(loc = 'lower right')

plt.savefig('simple-plot.png', dpi = 150) # dpi is for resolution of plot

# plt.savefig('C://Desktop/simple-plot.png', dpi=150) (if I want to specify the file)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha =.5)
plt.hlines(0, date[0], date[-1], linestyle = '--', colors = 'grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 50, 300)
y = np.exp(-x/10) * np.sin(x)


plt.plot(x, y)
plt.xlabel('x', size = 15, c = 'darkblue')
plt.ylabel('y', size = 15, c = 'darkblue', rotation = 0)
plt.title(r'$ y = e^{-x/10} \sin(x)$', size = 20)
plt.grid(alpha = 0.25)
plt.ylim([-1, 1])
plt.show()
```
